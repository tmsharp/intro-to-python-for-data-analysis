# Intro to Python for Data Analysis (Part 3): Data Visualization

```python
import pandas as pd
df = pd.read_csv("data/iris.csv")
df.sample(5)
```

```python
colors = ['red', 'orange', 'blue']
classes = df['class'].unique()
color_map = dict(zip(classes, colors))
print("Our color map:", color_map)

df['color'] = df['class'].apply(lambda x: color_map[x])
df.sample(5)
```

## 1. Matplotlib

```python
from IPython.display import Image; Image(filename='img/matplotlib.png')
```

https://matplotlib.org/tutorials/introductory/pyplot.html

```python
import matplotlib.pyplot as plt
```

```python
# a simple scatterplot using matplotlib 
plt.scatter(df['sepal_length'], y=df['petal_length'], c=df['color'])
```

```python
# scatter figure with axes and title 
plt.scatter(x=df['sepal_length'], y=df['petal_length'], c=df['color'])

plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.title('The Iris Dataset')

plt.show()
```

```python
# add legend 

# create subplots
fig, ax = plt.subplots()

# loop through each class, create a subplot
for cls in df['class'].unique():
    x = df[df['class'] == cls]['sepal_length']
    y = df[df['class'] == cls]['petal_length']
    ax.scatter(x=x, y=y, c=color_map[cls], label=cls)
    

# add axes labels, title, and legend
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.title('The Iris Dataset')
ax.legend()

# display 
plt.show()
```

```python
# save figure before displaying 


# create subplots
fig, ax = plt.subplots()

# loop through each class, create a subplot
for cls in df['class'].unique():
    x = df[df['class'] == cls]['sepal_length']
    y = df[df['class'] == cls]['petal_length']
    ax.scatter(x=x, y=y, c=color_map[cls], label=cls)
    

# add axes labels, title, and legend
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.title('The Iris Dataset')
ax.legend()

# save and display 
filename = "plots/iris_scatterplot"
plt.savefig(filename)
```

## 2. Seaborn 

```python
from IPython.display import Image; Image(filename='img/seaborn.png') 
```

http://seaborn.pydata.org/tutorial.html

```python
import seaborn as sns
```

```python
# use seaborn
sns.scatterplot(x="sepal_length", y="petal_length", hue="class", data=df)
```

```python
# set up custom color palette
colors = ['red', 'orange', 'blue']
my_palette = sns.color_palette(colors)

print("Color palette created!")
```

```python
# create plot
scatter = sns.scatterplot(x="sepal_length", y="petal_length", hue='class', palette=my_palette, data=df)

# set labels, title, and legend title
scatter.set(xlabel='Sepal Length (cm)', ylabel='Petal Length (cm)', title='The Iris Dataset - Using Seaborn!')

# save 
plt.savefig("plots/iris_scatterplot-seaborn")
```

#### Other plots with Seaborn

```python
# setting figure size
sns.set(rc={'figure.figsize':(11.7,8.27)})
```

```python
# bar chart 
bar = sns.barplot(x='class', y='petal_length', data=df, palette=my_palette)
bar.set(xlabel='Species', ylabel='Petal Length (cm)', title='Iris Bar Chart')
plt.show()
```

```python
sns_pairplot = sns.pairplot(df, hue='class', palette=my_palette, height=2.5)

# save 
plt.savefig("plots/iris_pairplot")
```
